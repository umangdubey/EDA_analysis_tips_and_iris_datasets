```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=sns.load_dataset("iris")
```

```python
df.shape
```

## univariate analysis

```python
df_s=df.loc[df['species']=='setosa']
```

```python
df_v=df.loc[df['species']=='virginica']
```

```python
df_ver=df.loc[df['species']=='versicolor']
```

```python
plt.plot(df_s['petal_length'])
```

```python
plt.plot(df_s['sepal_length'],np.zeros_like(df_s['sepal_length']),'o')
plt.plot(df_v['sepal_length'],np.zeros_like(df_v['sepal_length']),'o')
plt.plot(df_ver['sepal_length'],np.zeros_like(df_ver['sepal_length']),'o')
plt.xlabel('petal_length')
plt.show
```

## Bivariate analysis

```python
sns.FacetGrid(df,hue='species',size=5).map(plt.scatter,'sepal_length','sepal_width').add_legend();
plt.show()
```

## multivariate analysis

```python
sns.pairplot(df,hue='species',size=3)
```
